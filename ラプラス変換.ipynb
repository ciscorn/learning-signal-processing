# ラプラス変換

- フーリエ変換をより実用本位に改造したもの。因果性信号のみを対象とする。
- 微分方程式で記述されるシステムを代数的な演算で解析できるようにする。工学的に有力なツールであり、特にアナログ信号処理システムの解析には欠くことができない。
- 信号における微分積分が、周波数領域では乗算や除算になることを利用する。

## ラプラス変換の定義

フーリエ変換との違いは以下の通り：

- フーリエ変換の $j\omega$ の代わりに、任意の複素数 $s$ を使う。
- 因果性信号のみを対象とする。つまり、積分区間を $[0, \infty)$ とする。

したがって、以下のように定義される：

$$F(s) = \int_0^\infty f(t) e^{-st} ds $$

これは、以下のフーリエ変換と等しい。

$$F(s) = F(a + j\omega) = \int_{-\infty}^\infty f(t) {\bf 1}(t)e^{-at} e^{-j\omega t} ds $$

つまりラプラス変換は、**収束因子** $e^{-at}$ をかけた因果性信号 $f(t) {\bf 1}(t)e^{-at}$ のフーリエ変換と考えることができる。

ラプラス変換の主な特徴

- フーリエ変換は積分が収束しない信号に対して定義できない。それらの信号も、収束因子によって、ラプラス変換の対象となる。
- 因果性信号を前提に構築しているため、$t=0$における初期条件を与えて解析できる（過渡解析ができる）。

## ラプラス変換の性質

基本的にフーリエ変換と似ている。微分は少し違う（初期条件が付く）。

### 線形性

これは自明

### 拡大縮小

$$ f(at) \Longleftrightarrow \frac{1}{a} F \left(\frac{s}{a}\right) $$


### 遅延（時間推移）

$$ f(t - t_0) \Longleftrightarrow F(\omega) e^{-s t_0}$$

### 微分

$$ \frac{df(x)}{dt} \Longleftrightarrow s F(s) - f(0) $$

### 高階微分

$$ \frac{df^{(n)}(x)}{dt} \Longleftrightarrow s L\left[ f^{(n-1)}(t) \right] -f^{(n-1)}(0)  $$

つまり

$$ \frac{df''(x)}{dt} \Longleftrightarrow s^2 F(s) - sf(0) - f'(0) $$


$$ \frac{df^{n}(x)}{dt} \Longleftrightarrow s^n F(s) - s^{n-1}f(0) \cdots - f^{n-1}(0) $$

### 積分

$$\int_{-\infty}^t f(\tau) d\tau \Longleftrightarrow \frac{1}{s}F(s) $$



### 畳み込み

$f(t)$, $g(t)$, $f(t) \ast g(t)$ が因果性信号のとき:

$$ f(t) \ast g(t) \Longleftrightarrow F(s) G(s) $$

$$ f(t) g(t) \Longleftrightarrow F(s) \ast G(s) $$


## ラプラス逆変換

ラプラス変換から時間関数への変換をラプラス逆変換といい、次の複素積分で定義される。これをブロムウィッチ積分という。

$$f(t) = \frac{1}{2\pi j} \int_{\sigma-j\infty}^{\sigma+j\infty} F(s) e^{st} ds $$

ここで、$\sigma$は、ラプラス変換が収束する $a$ に対し、$\sigma > a$を満たすものとする。

（これをまともに解くには複素解析学の留数定理などが必要）

実用においては、ブロムウィッチ積分をせずに、ラプラス変換表を参照して逆変換することがほとんどである。

### 有理関数のラプラス逆変換

あとでみるように、線形微分方程式のラプラス変換は有理関数となるので、有理関数のラプラス逆変換は実用上重要である。

部分分数分解によって解く。（留数定理がわかれば、ブロムウィッチ積分のほうが簡単らしいが）

$$ F(s) = \frac{b_0 s^m + b_1 s^{m-1} + \cdots + b_m}{s^n + a_1 s^{n-1} + \cdots + a_n} \quad (n \geq m)$$

これは、分母多項式の根を $\lambda_i$ として、根に重複がないと仮定すると、次のように部分分数分解できる。

$$ F(s) = \frac{p_1}{s - \lambda_1} +  \frac{p_2}{s - \lambda_2} + \cdots + \frac{p_n}{s - \lambda_n}$$

以下のラプラス変換対を用いる：

$$ {\bf 1}(t) \Longleftrightarrow \frac{1}{s} $$

$$ e^{at} f(t) \Longleftrightarrow F(s+a) $$

すると、$f(t)$は以下であることがわかる

$$ f(t) = \sum p_i  e^{\lambda_i t} {\bf 1}(t)$$

多項式の係数が実数なので、$\lambda$が根の1つであるとき$\overline{\lambda}$も根である。したがって、上式の虚部は打ち消しあう。

$Re(\lambda_i)$がすべて負なら、$f(t)$が発散しないこともわかる。

$\lambda_i$が${k_i}$重根の場合は、以下を利用すれば：

$$ t^n \Longleftrightarrow \frac{n!}{s^{n+1}} $$

こうなる：

$$ f(t) = \sum p_i \frac{t^{k-1}}{(k-1)!} e^{\lambda_i t} {\bf 1}(t)$$
