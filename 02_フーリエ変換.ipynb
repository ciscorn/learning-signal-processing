# フーリエ変換

## 定義

フーリエ級数展開を、周期性のない一般の信号に拡張したものが、フーリエ変換である。

基本的にな考えは、フーリエ級数展開において、$lim_{\omega_0 \to 0}$ として、周期を無限大にすることである。

このとき、フーリエ級数展開の式にある $\omega_0 / 2\pi$ は、フーリエ逆変換側に回す。そうするとフーリエ級数の式は区分求積になり、さらに$lim_{\omega_0 \to 0}$によって、区分求積は積分となる（これが逆変換の式になる）。

フーリエ変換を以下のように定義する。

$$ F(\omega) = \int_{-\infty}^{+\infty} f(t) e^{-j\omega t} dt$$

フーリエ逆変換は以下のようになる。

$$ f(t) = \frac{1}{2\pi} \int_{-\infty}^{+\infty} F(\omega) e^{j\omega t} d\omega$$

係数$\frac{1}{2\pi}$の扱いについては様々な慣習がある。変換と逆変換の両方に $\frac{1}{\sqrt{2\pi}}$ を置く定義もある（この定義は、変換をユニタリにしたいときに便利である）。

信号とフーリエ変換の対を、フーリエ変換対と呼ぶ：

$$ f(t) \Longleftrightarrow F(\omega)$$

### n次元への拡張

ここでは扱わないが、フーリエ変換は任意のn次元空間にも拡張できる。拡張するときは $\omega$ と $x$ をベクトルにする。式中の$e^{-j \omega x}$ ではベクトル $\omega$ と $x$ のドット積を用いて、さらに先頭の係数を $1/(2\pi)^n$ にする。

なぜこのようにすると拡張できるのかは知らない。直感的には、例えば2次元の信号は、あらゆる方向への平面波の重ね合わせで表現できるということだが。

## スペクトル

- フーリエスペクトル $F(\omega)$
- 振幅スペクトル $|F(\omega)|$
- 位相スペクトル $arg F(\omega)$
- パワースペクトル $|F(\omega)|^2$ （振幅の2乗は電力の次元をもつ）

## 重要な具体例

### インパルス信号（デルタ関数）

$$\delta(t) \Longleftrightarrow 1$$

この逆変換を証明するのは難しそう。

### パルス信号（ボックス関数）

$$br(t) =
\begin{cases}
1 & -T/2 < t < T/2 \\
0 & \text{otherwise}
\end{cases}
$$

$$br(t) \Longleftrightarrow \frac{2\sin(\omega T/2)}{\omega} $$

これはsinc関数、つまり$sinc(x) = \frac{sin(x)}{x}$ の形をしている。つまり、

$$br(t) \Longleftrightarrow T sinc \left( \frac{\omega T}{2} \right) $$


## フーリエ変換の性質

### 線形性

これは自明

### 実信号において

$$ F(\omega) = \overline{F(-\omega)} $$

つまり実部は偶関数、虚部は奇関数となる。

振幅スペクトル $|F(\omega)|$ は偶関数、位相スペクトル $\arg F(\omega)$ は奇関数となる。

### 零位相特性（偶関数信号）

信号が偶関数 $f(t) = f(-t)$ のとき、フーリエスペクトルは零位相となる（虚数成分を持たない）。

### 双対性

フーリエ変換と逆変換とを比べると、符号の正負と係数$1/2\pi$が異なるだけで、それ以外の違いはない。

したがって例えば、信号がパルス関数のとき、そのフーリエ変換がsincであるのと同様に、フーリエ変換がパルス関数のとき、その信号はsincである。

$$ F(t) \Longleftrightarrow 2 \pi f(-\omega)$$

### 拡大縮小

時間軸方向に縮小（拡大）すれば、周波数軸方向で拡大（縮小）する。

$$ f(at) \Longleftrightarrow \frac{1}{a} F \left(\frac{\omega}{a}\right) $$

これは他の重要な意味ももっている。時間軸上で狭めれば周波数軸上では広がるし、周波数軸上で狭めれば時間軸では広がる。これを**信号とフーリエ変換の不確定性**という。信号を短い時間に集中させると広い帯域をもってしまう、狭い周波数帯域に集中させると長い時間の信号になってしまう。

### 時間推移

$$ f(t - t_0) \Longleftrightarrow F(\omega) e^{-j\omega t_0}$$

$e^{-j\omega t_0}$ は**遅延演算子**と呼ばれる。時間軸上での推移は、周波数領域で遅延演算子を掛けることに相当する。

つまり位相スペクトルに直線位相が加わる：

$$arg F(\omega) -\omega t_0$$

振幅スペクトルは変化しない。

### 周波数推移

周波数軸上での推移も同様である。

$$ f(t)e^{j\omega_0 t} \Longleftrightarrow F(\omega - \omega_0) $$

これを用いて、変調信号が導ける。

$$f(t)\cos(\omega t) \Longleftrightarrow \frac{F(\omega - \omega_0) + F(\omega + \omega_0)}{2}$$


### 時間の反転

$$ f(-t) \Longleftrightarrow F(-\omega) $$

実信号の場合は、$F(-\omega) = \overline{F(\omega)}$ なので、

$$ f(-t) \Longleftrightarrow \overline{F(\omega)} $$

### 微分

$$ \frac{df(x)}{dt} \Longleftrightarrow j\omega F(\omega) $$

### 積分

$$\int_{-\infty}^t f(\tau) d\tau \Longleftrightarrow \frac{1}{j\omega}F(\omega) $$

### 畳み込み

$$ f(t) \ast g(t) \Longleftrightarrow F(\omega) G(\omega) $$

$$ f(t) g(t) \Longleftrightarrow F(\omega) \ast G(\omega) $$

証明は容易

### 信号のエネルギーとパワースペクトルの関係（パーセバルの定理）

信号 $f(t)$ のエネルギーを以下のように定義する。これは電位差が $f(t)$ のときに、1[Ω]の負荷で消費される電力量である。

$$E = \int_{-\infty}^{\infty}|f(t)|^2 dt$$

パワースペクトルとの関係は以下の通りとなる。これをParseval's theoremという。

$$\int_{-\infty}^{\infty}|f(t)|^2 dt = \frac{1}{2\pi} \int_{-\infty}^{\infty}|F(\omega)|^2 d\omega$$

（証明方法は知らない。ただ、フーリエ変換が直交変換（ユニタリー変換）であることを踏まえて、線形代数の知見を使えば、これは当然といえば当然である。）

（ちなみに、フーリエ変換・逆変換の係数として、$1/\sqrt{2\pi}$を使えば、完全にユニタリーになり、上式の$1/2\pi$も消える。）

この関係により、パワースペクトルは、信号のエネルギーのうち、特定の周波数帯域がもつエネルギーを表していると考えられる。
